# KNN Implementation using Euclidian Distance

```python
def custom_KNN(l_metrics):
    
    l_e_distances    = []
    l_len           = len(data.uid)
    
    

    l_genere            = data["genre"].values
    l_given_genres      = len(l_metrics[0])
    
    l_uids             = data["uid"].values
    l_titles           = data["title"].values
    
    l_popularity       = data["popularity"].values
    l_episodes         = data["episodes"].values
    l_rank             = data["ranked"].values
    
    for i in range(l_len):
        l_cnt_genere  = 0
        l_pop         = 0
        l_epi         = 0
        l_rank_1      = 1
        l_e_distance  = 0
        
        #checking metric genere
        for j in l_metrics[0]:
            if j in l_genere[i]:
                l_cnt_genere = l_cnt_genere + 1
        
        #checking metric popularity(if popularity is below metioned treating it as 1 else treating it as 0)
        if l_popularity[i] < l_metrics[1]:
            l_pop = 1
        
        #checking metric episodes(checking distance from given number of episodes to number of episodes of that anime)
        l_epi = l_episodes[i]
        
        #checing metric rank(if the anime ranks below the mentioned rank, then it is treated as 0 or else it is 1)
        if l_rank[i] < l_metrics[3]:
            l_rank_1 = 0
            
        l_e_distance =  (((l_given_genres-l_cnt_genere)**2) + ((l_pop)**2) + ((l_metrics[2]-l_epi)**2) + ((l_rank_1)**2))**0.5
        l_e_distances.append([i,l_uids[i],l_titles[i],l_e_distance])
            
    return l_e_distances
```

```python
def KNN(l_metrics,k = 10):
    
    l_e_distances = custom_KNN(l_metrics)
    
    l_e_distances.sort(key=lambda row: (row[3]))
    
    #print(l_e_distances)
    
    for i in range(k):
        print(l_e_distances[i][2])
    
```

```python
#input metrics Genre, pop(above that number is liked), Episodes(episodes approx to that number), rank(below that num is liked)
l_metrics = [["Adventure","Drama"],34,1000,100]
KNN(l_metrics, 3)
```
